```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
model = tf.keras.models.load_model('model.h5')
```

```python
def predict(image_28by28by1):
  img=image_28by28by1
  img=img.reshape(-1,28,28,1)
  ind=np.argmax(model.predict(img))
  return chr(65+ind)
```

```python
import cv2 as cv
```

```python
img=cv.imread('try1.jpg')
```

```python
plt.imshow(img)
```

```python
img=img[1200:2000, 500:1400]
```

```python
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
gray=cv.bitwise_not(gray)
```

```python
for i in range(len(gray)):
  for j in range(len(gray[i])):
    if(gray[i][j]>100):gray[i][j]=255
    else: gray[i][j]=0
```

```python
plt.imshow(gray)
```

```python
stretch_near = cv.resize(gray, (28,28))
```

```python
plt.imshow(stretch_near)
```

```python
predict(stretch_near)     #correct
```

```python
img=cv.imread('M.jpg')
img=img[1000:1700, 750:1450]
plt.imshow(img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray=cv.bitwise_not(gray)
for i in range(len(gray)):
  for j in range(len(gray[i])):
    if(gray[i][j]>100):gray[i][j]=255
    else: gray[i][j]=0
fin_img = cv.resize(gray, (28,28))
predict(fin_img)  #correct
```
